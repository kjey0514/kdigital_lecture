```python
td=pd.read_csv("titanic_train.csv")
td
```

```python
values = {'age': td['Age'].fillna(td['Age'].mean(), inplace=True), 'Cabin': 'N', 'Embarked': 'N'}
td.fillna(values)
td
```

```python
# survived_td=td[td['Survived']>0]
td.groupby('Sex')['Survived'].sum()
```

```python
td.groupby('Pclass')['Survived'].sum()
```

```python
age_group = {'10대': td[td['Age']<20]['Survived'].sum(),
             '20대': td[td['Age']<30]['Survived'].sum(),
             '30대': td[td['Age']<40]['Survived'].sum(),
             '40대': td[td['Age']<50]['Survived'].sum(),
             '50대': td[td['Age']<60]['Survived'].sum(),
             '60대': td[td['Age']<70]['Survived'].sum(),
             '노인': td[td['Age']>=70]['Survived'].sum(),
             '유아':td[td['Age']<10]['Survived'].sum()
             }
age_group
```

```python
td['Age_Range']= td['Age'].apply(lambda x: '유아' if x< 10 else '10대' if x<20 else '20대' if x<30 
                                else '30대' if x<40 else '40대' if x<50 else '50대' if x<60
                                 else '60대' if x<70 else '노인' )
td
```
